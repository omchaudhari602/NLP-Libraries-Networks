| **Customization** | High flexibility for custom experimentation | Strong customization via pipeline components, though more structured |
| **Best use cases** | Academic work, prototyping NLP concepts, teaching | Production systems, APIs, high-throughput NLP services |

4: Describe the architecture and functioning of a Recurrent Natural Network
(RNN).

ans:

# Architecture and Functioning of a Recurrent Neural Network (RNN)


A Recurrent Neural Network (RNN) is a type of neural network designed to handle
sequential data such as text, time series, or speech. Unlike traditional neural
networks, RNNs have connections that form loops, allowing information to persist
across time steps.

Architecture of RNN:
1. Input Layer:
   - Takes sequence data as input (e.g., words in a sentence one by one).

2. Hidden Layer:
   - The core of the RNN where computation happens.
   - It receives both the current input and the previous hidden state.
   - This creates a "memory" of past information.

3. Output Layer:
   - Produces output at each time step (e.g., next word prediction, classification).

Key Components:
- xt: Input at time step t
- ht: Hidden state at time step t (memory)
- ht-1: Previous hidden state
- yt: Output at time step t

Mathematical Representation:
ht = f(Wx * xt + Wh * ht-1 + b)
yt = g(Why * ht + by)

Where:
- Wx, Wh, Why are weight matrices
- b, by are bias terms
- f is usually tanh or ReLU
- g is activation function like softmax (for classification)

Functioning of RNN:
1. The network processes input sequentially (one element at a time).
2. At each step, it updates the hidden state using current input and previous state.
3. The hidden state carries information forward through the sequence.
4. Output is generated at each step or only at the final step depending on the task.

Advantages:
- Captures temporal dependencies in sequential data
- Useful for tasks like language modeling, machine translation, and speech recognition

Limitations:
- Vanishing and exploding gradient problems
- Difficulty learning long-term dependencies

Solution Extensions:
- LSTM (Long Short-Term Memory)
- GRU (Gated Recurrent Unit)

Summary:
RNNs are powerful for sequence modeling because they use feedback connections
to retain past information, making them suitable for many NLP and time-dependent tasks.

5: What is the key difference between LSTM and GRU networks in NLP
applications?
ans:

# Key Difference Between LSTM and GRU in NLP Applications


LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) are advanced types
of Recurrent Neural Networks (RNNs) designed to solve the problem of learning
long-term dependencies in sequential data.

Key Differences:

1. Structure:
   - LSTM has a more complex structure with THREE gates:
     * Input Gate
     * Forget Gate
     * Output Gate
     * It also has a separate cell state (Ct) for memory.

   - GRU has a simpler structure with TWO gates:
     * Update Gate
     * Reset Gate
     * It combines hidden state and cell state into one.

2. Complexity:
   - LSTM is more complex and has more parameters.
   - GRU is simpler and computationally faster.

3. Memory Handling:
   - LSTM explicitly controls memory using a separate cell state.
   - GRU directly updates the hidden state without a separate memory cell.

4. Training Speed:
   - LSTM takes longer to train due to more parameters.
   - GRU trains faster and requires less data.

5. Performance:
   - LSTM performs better on tasks requiring long-term dependencies.
   - GRU often performs similarly and sometimes better on smaller datasets.

Summary:
The main difference is that LSTM has a more detailed memory control mechanism
with three gates and a cell state, while GRU is a simplified version with fewer
gates, making it faster and more efficient in many NLP tasks.

7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:

“Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.”

(Include your Python code and output in the code box below.)


```python
# Install NLTK (run once in Colab)
!pip install nltk

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required resources
nltk.download('punkt')
nltk.download('punkt_tab') # Added to address the LookupError

# Given paragraph
text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# Tokenization
tokens = word_tokenize(text)

# Convert to lowercase and filter only alphabetic words
words = [word.lower() for word in tokens if word.isalpha()]

# Frequency Distribution
freq_dist = FreqDist(words)

# Print tokens
print("Tokens:\n", words)

# Print frequency distribution
print("\nFrequency Distribution:")
for word, freq in freq_dist.items():
    print(word, ":", freq)
```

8: Implement a basic LSTM model in Keras for a text classification task using
the following dummy dataset. Your model should classify sentences as either positive
(1) or negative (0).


# Dataset
texts = [
“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive
]
labels = [1, 1, 0, 0, 1]
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
this data. You may use Keras with TensorFlow backend.

(Include your Python code and output in the code box below.)


```python
# Install TensorFlow (if not already installed)
!pip install tensorflow

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",  # Positive
    "This is an amazing experience",  # Positive
    "I hate waiting in line",  # Negative
    "This is the worst service",  # Negative
    "Absolutely fantastic!"  # Positive
]

labels = np.array([1, 1, 0, 0, 1])

# Tokenization
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# Padding
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# Build LSTM Model
model = Sequential([
    Embedding(input_dim=1000, output_dim=16, input_length=max_length),
    LSTM(32),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(padded_sequences, labels, epochs=20, verbose=1)

# Test prediction
test_text = ["I really love this service"]
test_seq = tokenizer.texts_to_sequences(test_text)
test_pad = pad_sequences(test_seq, maxlen=max_length, padding='post')

prediction = model.predict(test_pad)

print("\nTest Sentence:", test_text[0])
print("Predicted Sentiment (1=Positive, 0=Negative):", int(prediction[0][0] > 0.5))
print("Prediction Score:", prediction[0][0])
```

9. Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:

“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.”

Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.


```python
# Install spaCy
!pip install spacy

# Download English model
!python -m spacy download en_core_web_sm

import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Given paragraph
text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

# Process text
doc = nlp(text)

# Print Tokens and Lemmas
print("Tokens and Lemmas:\n")
for token in doc:
    print(f"{token.text} --> {token.lemma_}")

# Print Named Entities
print("\nNamed Entities:\n")
for ent in doc.ents:
    print(f"{ent.text} --> {ent.label_}")
```

- Padding: Ensure all sequences have the same length for model input.

- Privacy: Ensure user data is secure and protected.
